A qubit can take two values, either 0 or 1.
We can express a qubit state as

$$
\mid 0 \rangle \, or \\
\mid 1 \rangle.
$$

These states are visualized as the north and south poles a unit sphere called the $\textit{Bloch sphere}$.

In the following, we will draw a Block sphere, which has a radius of unit 1. 
The state $\mid 0 \rangle$ corresponds to the point (0, 0, 1), where we have (x, y, z) coordinates.

```python
from qiskit.visualization import plot_bloch_vector
%matplotlib inline

plot_bloch_vector([0, 0, 1])
```

In a similar fashion, the state $\mid 1 \rangle$ corresponds to the point (0, 0, -1) in this Bloch sphere.

```python
plot_bloch_vector([0, 0, -1])
```

### Superposition

We can combine the states to create a superposition. For instance, consider a state that has halfway between the north and south poles of the Bloch sphere:

$$
\frac{1}{\sqrt{2}} (\mid 0 \rangle + \mid 1 \rangle).
$$

This is at the (1, 0, 0) coordinate.

```python
plot_bloch_vector([1, 0, 0])
```

In y-axis we use, a phase of $i = \sqrt{-1}$

The six states in the equator of the Bloch sphere are denoted with special names:

$$
\mid + \rangle = \frac{1}{\sqrt{2}} (\mid 0 \rangle + \mid 1 \rangle), \\
\mid - \rangle = \frac{1}{\sqrt{2}} (\mid 0 \rangle - \mid 1 \rangle), \\
\mid i \rangle = \frac{1}{\sqrt{2}} (\mid 0 \rangle + i \mid 1 \rangle), \\
\mid -i \rangle = \frac{1}{\sqrt{2}} (\mid 0 \rangle - i \mid 1 \rangle), \\
$$

We can measure only a single qubit. The result is always either 0 or 1 with some probability and not a superposition of 0 and 1.

Consider the following qubit:

$$
\frac{1}{\sqrt{2}} (\mid 0 \rangle + \exp^{i \pi /6}\mid 1 \rangle).
$$

We can rewrite this qubit in the following format:

$$
\alpha \mid 0 \rangle + \beta \mid 1 \rangle),
$$
where
$$
\alpha  = \frac{1}{\sqrt{2}} \\
\beta =  \frac{1}{\sqrt{2}} \exp^{i \pi /6}.
$$


These are the coefficients of $\mid 0 \rangle$ and $\mid 1 \rangle$, respectively. The coefficients are called amplitudes. The probability is computed by the norm-square of the amplitude.

Therefore, for the above qubit, the probability of $\mid 0 \rangle$ is

$$
\left| \frac{1}{\sqrt{2}} \right|^2 = 0.5, 
$$

and the probability of $\mid 1 \rangle$ is

$$
\left| \frac{\exp^{i \pi /6}}{\sqrt{2}} \right|^2 =  \frac{\exp^{i \pi /6}}{\sqrt{2}} \frac{\exp^{-i \pi /6}}{\sqrt{2}} = \frac{\exp^0}{2} = 0.5. 
$$




We can measure in three different phases:

- $\{ \mid 0 \rangle, \mid 1 \rangle\}$ is called the Z-basis.
- $\{ \mid + \rangle, \mid - \rangle\}$ is called the X-basis.
- $\{ \mid i \rangle, \mid -i \rangle\}$ is called the Y-basis.


### Generic quantum state

A generic quantum state or generic qubit can be written as

$$
\mid \psi \rangle = \alpha \mid 0 \rangle + \beta \mid 1 \rangle.
$$

Here, $\mid \alpha\mid^2 + \mid \beta \mid^2 = 1$. To determine the location of this qubit, we can first parameterize $\alpha$ and $\beta$ with $\theta$ and $\phi$ as 

$$
\alpha = \cos(\frac{\theta}{2}), \\
\beta = \exp^{i\phi} \sin(\frac{\theta}{2}).
$$


Therefore, we can express the parameterized generic state as

$$
\mid \psi \rangle = \cos(\frac{\theta}{2}) \mid 0 \rangle + \exp^{i\phi} \sin(\frac{\theta}{2}) \mid 1 \rangle.
$$
